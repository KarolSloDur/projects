```python
from sklearn.model_selection import train_test_split
df = pd.read_csv('housing-classification-iter6.csv').set_index('Id')
test_data = pd.read_csv('test-housing-classification.csv').set_index('Id')
```

```python
test_data
```

```python
X=df
y=X.pop('Expensive')
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
import numpy as np
from sklearn.impute import SimpleImputer
```

```python
X_train_cat = X_train.select_dtypes(exclude="number")
X_test_cat=X_test.select_dtypes(exclude='number')
test_data_cat=test_data.select_dtypes(exclude='number')
```

```python
X_train_num=X_train.select_dtypes(include="number")
X_test_num=X_test.select_dtypes(include="number")
test_data_num=test_data.select_dtypes(include='number')
```

```python
imp = SimpleImputer(strategy='median').set_output(transform='pandas')
```

```python
imp.fit(X_train_num)
```

```python
X_train_num_imp = imp.transform(X_train_num)
```

```python
X_test_num_imp = imp.transform(X_test_num)
test_data_num_imp = imp.transform(test_data_num)
test_data_num_imp
```

```python
test_data_num_imp.isnull().sum()
```

```python

# defining the imputer to use "N_A" as replacement value
cat_imputer = SimpleImputer(strategy="constant", 
                            fill_value="N_A").set_output(transform='pandas')

# fitting and transforming
X_train_cat_imputed = cat_imputer.fit_transform(X_train_cat)

X_train_cat_imputed.head()
X_test_cat_imputed = cat_imputer.transform(X_test_cat)
test_data_cat_imputed = cat_imputer.transform(test_data_cat)
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python

my_onehot = OneHotEncoder(sparse_output=False, handle_unknown="ignore").set_output(transform='pandas')

# fit and trnsform train data
X_train_cat_imp_onehot =my_onehot.fit_transform(X_train_cat_imputed)

# transform
X_test_cat_imp_onehot = my_onehot.transform(X_test_cat_imputed)
test_data_cat_imp_onehot = my_onehot.transform(test_data_cat_imputed)
```

```python
X_train_conc = pd.concat([X_train_cat_imp_onehot, X_train_num_imp], axis=1)
X_test_conc=pd.concat([X_test_cat_imp_onehot, X_test_num_imp], axis=1)
test_data_conc=pd.concat([test_data_cat_imp_onehot, test_data_num_imp], axis=1)
```

```python
X_test_conc.head()
```

```python
test_data_conc.head()
```

```python
from sklearn import preprocessing
scaler = preprocessing.StandardScaler().set_output(transform='pandas')
X_train_conc_norm=scaler.fit_transform(X_train_conc)
X_test_cont_norm=scaler.transform(X_test_conc)
test_data_cont_norm=scaler.transform(test_data_conc)
```

```python
from sklearn.tree import DecisionTreeClassifier
tree_model = DecisionTreeClassifier(criterion='gini', min_samples_leaf=3)
tree_model.fit(X_train_conc_norm, y_train)
```

```python
y_train_pred = tree_model.predict(X_train_conc_norm)
```

```python
y_test_pred=tree_model.predict(X_test_cont_norm)
```

```python
y_test_data = tree_model.predict(test_data_cont_norm)
```

```python
from sklearn.metrics import accuracy_score
train_accuracy = accuracy_score(y_train, y_train_pred)
print("Train accuracy:", train_accuracy)
```

```python
test_accuracy=accuracy_score(y_test, y_test_pred)
print("Test accuracy:", test_accuracy)
```
