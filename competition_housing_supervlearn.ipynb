```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import RandomizedSearchCV
from sklearn.pipeline import make_pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import MaxAbsScaler
from sklearn.decomposition import TruncatedSVD
from sklearn import linear_model
from sklearn.feature_selection import SelectFromModel
from sklearn.model_selection import RandomizedSearchCV
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn import linear_model
from sklearn.linear_model import PoissonRegressor
house_all = pd.read_csv('train.csv')
house_test=pd.read_csv('test.csv')
house_all.head()
```

```python
X = house_all.drop(columns=['SalePrice'])
y = house_all['SalePrice']

# Split the dataset into train and test sets
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size=0.05, random_state=42)
```

Preprocessor

```python

# Identify the numerical and categorical columns
num_cols = X.select_dtypes(include='number')
cat_cols = X.select_dtypes(exclude='number')
categorical_pipe = make_pipeline(SimpleImputer(strategy="constant", fill_value="N_A"),OneHotEncoder(handle_unknown='ignore', drop='first'))
num_cols=num_cols.drop('Id', axis=1 )
# Create transformers for numerical and categorical features
num_pipe = make_pipeline(SimpleImputer())


# Create a ColumnTransformer to apply the transformers to the appropriate columns
preprocessor = ColumnTransformer(
    transformers=[
        ('num', num_pipe, num_cols.columns),
        ('cat', categorical_pipe, cat_cols.columns)
    ])
```

DecisionTreeRegressor

```python

# Combine the preprocessor and the model into a single pipeline
pipeline = Pipeline([
    ('preprocessor', preprocessor),('scale',MaxAbsScaler()), ('sele',SelectFromModel(DecisionTreeRegressor())),
    ('model', DecisionTreeRegressor())])


# Fit the pipeline to the training data
pipeline.fit(X_train, y_train)

# Make predictions
y_train_pred_all = pipeline.predict(X_train)
y_test_pred_all = pipeline.predict(X_test)
y_test_pred_test = pipeline.predict(house_test)
```

```python

from sklearn.metrics import r2_score
performances = pd.DataFrame(columns=["train", "test"])

performances.loc["decision_tree", "train"] = r2_score(y_train, y_train_pred_all)
performances.loc["decision_tree", "test"] = r2_score(y_test, y_test_pred_all)
performances
```
